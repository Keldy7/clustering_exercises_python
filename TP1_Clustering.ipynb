```python
# Importation des librairies

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris # Chargement du dataset Iris
from sklearn.decomposition import PCA # DÃ©composition en analyse en composantes principales
from sklearn.preprocessing import StandardScaler
```

```python
# Chargement et affichage   
iris = load_iris()
```

```python
# SÃ©paration des caractÃ©ristiques (X) et des Ã©tiquettes (y)
X = iris.data 
y = iris.target
print(f"Nombre d'observations : {X.shape[0]}")
print(f"Nombre de variables : {X.shape[1]}")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
```

```python
# Standardisation des donnÃ©es
X_scaled = StandardScaler().fit_transform(X)
```

```python
# Initialisation de l'ACP avec 2 composantes principales et transformation des donÃ©es
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_scaled)
```

```python
# Visualisation des rÃ©sultats Ã  l'aide d'un nuage de points
plt.figure()
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('PrÃ©miÃ¨re composante principale ')
plt.ylabel('DeuxiÃ¨me composante principale')
plt.title("VisualiSation de l'ACP")
plt.colorbar()
plt.show()
```
